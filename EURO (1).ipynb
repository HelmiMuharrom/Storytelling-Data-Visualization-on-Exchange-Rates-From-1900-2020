# What others?
Another aspect that we need to consider is the audience's direction of reading. As they scroll down through our article, they will read (and look at) our graph from top to bottom. More specifically, people usually start from top left and follow a zigzag pattern until they reach bottom right.

One problem with our graph is that the tick labels are located at the bottom. People will immediately see the country names, the bars, but they may get confused about the quantities.

So we do change
1. Color the horizontal bar plot using the HEX color code #af0b1e.
2. Move the x-tick labels on top and make sure no ticks are visible.
3. Color the x-tick labels in grey.

```python
top20_deathtoll = pd.read_csv('top20_deathtoll.csv')

fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'],
        height=0.45, color='#af0b1e')
for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)
ax.set_xticks([0, 150000, 300000])
#giving the title at the top
ax.xaxis.tick_top()
ax.tick_params(top=False, left=False)
ax.tick_params(axis='x', colors='grey')
ax.text(x=-80000, y=23.5,
        s='The Death Toll Worldwide Is 1.5M+',
        weight='bold', size=17)
ax.text(x=-80000, y=22.5,
        s='Top 20 countries by death toll (December 2020)',
        size=12)


# *** *** *** *** *** *** *** *** *** ***

# Solution Code
ax.set_xticklabels(['0', '150,000', '300,000'])
ax.set_yticklabels([]) # an empty list removes the labels
country_names = top20_deathtoll['Country_Other']
for i, country in zip(range(20), country_names):
    ax.text(x=-80000, y=i-0.15, s=country)
ax.axvline(x=150000, ymin=0.045, c='grey',
           alpha=0.5)

plt.show()
```

# Data Cleaning Procces<a class="anchor" id="section_2_1"></a>

Our focus in the guided part of the project will be on the exchange rate between the euro and the American dollar. If you want to explore a different currency after finishing the guided part, the code we use for cleaning will come in handy.

Below, we do the following:

1. We rename the **[US dollar ]** and **Period\Unit**: columns to something easier to type — US_dollar and Time.
2. We change the Time column to a datetime data type.
3. We sort the values by Time in ascending order.
4. We reset the index (and drop the initial index).

```python
exchange_rates.rename(columns={'[Indonesian rupiah ]': 'Idr_rupiah',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_idr = exchange_rates[['Time','Idr_rupiah']].copy()
exchange_rates['Idr_rupiah'].value_counts()# 62 '-' characters

#Drop all the rows where the - character appears in the Idr_rupiah column.
euro_to_idr = euro_to_idr[euro_to_idr['Idr_rupiah'] != '-']

#Convert the Idr_rupiah column to a float data type.
euro_to_idr['Idr_rupiah'] = euro_to_idr['Idr_rupiah'].astype(float)
euro_to_idr['Idr_rupiah'].info()
```

# For Better Understanding Our Dataset
 ## About The Rolling Mean
 The rolling mean, also known as moving average, is a technique used to smooth out a time series by averaging data over a fixed window size. In other words, instead of plotting the raw data, we plot a line that represents the average of the data points in a given window. This is useful because it helps to remove the noise or fluctuations that occur at a daily or hourly level, and to highlight the long-term trends in the data.
 
  method is applied to the daily exchange rate values with various rolling window sizes (e.g., 7 days, 30 days, 365 days), which calculates the average value of the data points within the window. The resulting rolling mean values are then plotted instead of the raw data. As the window size increases, the line becomes smoother and highlights the long-term trends in the data.

# Lets find out
Calculate the rolling means for the IDR_RUPIAh column using a moving window of 30 days. Add the rolling means to a new column named rolling_mean.

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_idr['Time'], euro_to_idr['Idr_rupiah'])
plt.title('Original values', weight='bold')

# Create subplots for each rolling mean window size
for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_idr['Time'],
             euro_to_idr['Idr_rupiah'].rolling(rolling_mean).mean())
    #Giving Team an title
    plt.title('Rolling Window:' + str(rolling_mean), weight = 'bold')
    plt.tight_layout() # Auto-adjusts the padding between subplots
```

```python
euro_to_idr['rolling_mean'] = euro_to_idr['Idr_rupiah'].rolling(30).mean()
euro_to_idr
```
